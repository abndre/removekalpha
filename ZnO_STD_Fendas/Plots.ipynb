```python
import matplotlib.pyplot as plt
import numpy as np
import os
```

```python
print(os.listdir("../ZnO_STD_Fendas"))
```

```python
listfiles = os.listdir("../ZnO_STD_Fendas")
listafiles_new =[]
for i in listfiles:
    if '.xy' in i:
        listafiles_new.append(i)
        
```

```python
namefile = 'ZnO_STD_202_F_XY.xy'
x,y= np.loadtxt(namefile, unpack= True)
```

```python
plt.plot(x,y)
plt.show()
```

```python
def normalizar (y):
    maxi = max(y)
    for key,value in enumerate(y):
        y[key]=value/maxi
    return y
        
```

```python
for i in listafiles_new:
    if '101' in i:
        x,y= np.loadtxt(i, unpack= True)
        y=normalizar(y)
        #print(i)
        plt.plot(x,y)

plt.xlabel('Position ($2\Theta$)')
plt.ylabel('Intensity Normalized (U.A.)')
plt.show()
```
